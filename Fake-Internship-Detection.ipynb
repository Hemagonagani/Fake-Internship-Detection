```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

data = {
    'Platform': ['LinkedIn','Unknown','Website','LinkedIn','Unknown','Website','LinkedIn','Unknown'],
    'Fee': ['No','Yes','Yes','No','Yes','Yes','No','Yes'],
    'Certificate': ['Yes','Yes','No','Yes','No','Yes','Yes','No'],
    'Duration_Weeks': [8,4,6,10,3,5,12,2],
    'Placement_Promise': ['No','Yes','No','No','Yes','Yes','No','Yes'],
    'Label': ['Real','Fake','Fake','Real','Fake','Fake','Real','Fake']
}

df = pd.DataFrame(data)
df
```

```python
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])

X = df.drop('Label', axis=1)
y = df['Label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
# Example internship details
# Platform, Fee, Certificate, Duration, Placement
sample = [[1,1,0,3,1]]
result = model.predict(sample)

print("Prediction:", "Fake" if result[0]==1 else "Real")
```
